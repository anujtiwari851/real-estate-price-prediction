```python
import pulp
import numpy_financial as npf

# --- 1. Define Problem Data and Parameters ---

# Seasons and time blocks
seasons = ["Monsoon", "Dry"]
days = {"Monsoon": 122, "Dry": 243}
time_blocks = range(5)

# Household electricity demand (kWh)
demand = {
    "Monsoon": [0.8, 1.3, 1.1, 2.6, 0.5],
    "Dry":     [0.8, 1.6, 1.3, 2.4, 0.5]
}

# Solar generation per kWp installed (kWh/kWp) - Year 1 baseline
generation = {
    "Monsoon": [0.2, 1.5, 1.0, 0.0, 0.0],
    "Dry":     [0.7, 4.0, 3.0, 0.0, 0.0]
}

# Financial and technical parameters
discount_rate = 0.08
pv_capex = 65000
bess_capex = 28000
pv_area_per_kwp = 100
grid_buy_price = 9
grid_sell_price = 5
bess_efficiency = 0.90
initial_budget = 300000
max_rooftop_area = 500
degradation_rate = 0.005

# --- 2. Step 1: Solve a Simple LP to Find the Optimal System Size ---
# The most profitable strategy is to maximize the PV size within the budget.
model = pulp.LpProblem("Home_Energy_System_Design", pulp.LpMaximize)

# Define decision variables
C_PV = pulp.LpVariable("C_PV", lowBound=0)
C_BESS = pulp.LpVariable("C_BESS", lowBound=0)

# The objective is simply to install as much PV as possible.
model += C_PV, "Maximize_PV_Size"

# Define constraints
model += (pv_capex * C_PV + bess_capex * C_BESS) <= initial_budget, "Budget_Constraint"
model += pv_area_per_kwp * C_PV <= max_rooftop_area, "Rooftop_Area_Constraint"

# Solve for the optimal size
model.solve()

# Store the optimal system size
optimal_C_PV = C_PV.varValue
optimal_C_BESS = C_BESS.varValue

# --- 3. Step 2: Calculate the True NPV for the Optimal System ---
final_npv_cost = 0
if optimal_C_PV is not None:
    # Calculate initial investment (a positive cost at Year 0)
    initial_investment = pv_capex * optimal_C_PV + bess_capex * optimal_C_BESS

    # Calculate the present value of the BESS replacement cost at the start of year 11 (end of year 10)
    bess_replacement_pv = (bess_capex * optimal_C_BESS) / (1 + discount_rate)**10

    # Create lists of annual O&M costs and grid revenues over 25 years
    annual_om_costs = []
    annual_grid_revenues = []
    for year in range(1, 26):
        # O&M cost for the year
        annual_om_cost = 0.01 * initial_investment
        annual_om_costs.append(annual_om_cost)

        # Calculate annual grid revenue, applying this year's degradation
        degradation_factor = (1 - degradation_rate)**(year - 1)
        annual_net_grid_revenue = 0
        for s in seasons:
            for t in time_blocks:
                solar_gen = generation[s][t] * optimal_C_PV * degradation_factor
                load = demand[s][t]
                net_energy = solar_gen - load
                
                # Since BESS is 0, surplus is exported, deficit is imported
                if net_energy > 0:
                    annual_net_grid_revenue += days[s] * net_energy * grid_sell_price
                else:
                    annual_net_grid_revenue += days[s] * net_energy * grid_buy_price

        annual_grid_revenues.append(annual_net_grid_revenue)

    # Calculate the NPV of the future cost and revenue streams
    npv_of_om_costs = npf.npv(discount_rate, annual_om_costs)
    npv_of_grid_revenues = npf.npv(discount_rate, annual_grid_revenues)
    
    # Total NPV Cost = All costs (initial + future) - All future revenues
    final_npv_cost = initial_investment + bess_replacement_pv + npv_of_om_costs - npv_of_grid_revenues

# Display the Correct Optimal Solution
print(f"Size of the solar PV system: {optimal_C_PV:.2f} kWp")
print(f"Size of the BESS: {optimal_C_BESS:.2f} kWh")
print(f"Final 25-year Net Present Value of the total cost: ₹ {final_npv_cost:,.2f}")
```

```python
import pandas as pd

# --- System & Financial Parameters ---
optimal_C_PV = 4.62  # kWp
initial_investment = optimal_C_PV * 65000 # Cost of PV system [cite: 37]
degradation_rate = 0.005 # [cite: 42]
grid_buy_price = 9      # ₹/kWh 
grid_sell_price = 5   # ₹/kWh 

# --- Energy Profile Data ---
seasons = ["Monsoon", "Dry"]
days = {"Monsoon": 122, "Dry": 243} # [cite: 30]
demand = { # [cite: 33]
    "Monsoon": [0.8, 1.3, 1.1, 2.6, 0.5],
    "Dry":     [0.8, 1.6, 1.3, 2.4, 0.5]
}
generation = { # [cite: 35]
    "Monsoon": [0.2, 1.5, 1.0, 0.0, 0.0],
    "Dry":     [0.7, 4.0, 3.0, 0.0, 0.0]
}

# --- Baseline "Grid-Only" Calculation ---
annual_demand_kwh = sum(days[s] * sum(demand[s]) for s in seasons)
baseline_annual_bill = annual_demand_kwh * grid_buy_price # 

# --- Year-by-Year Calculation with PV System ---
results = []
for year in range(1, 26):
    degradation_factor = (1 - degradation_rate)**(year - 1)
    
    annual_generation_kwh = 0
    annual_grid_imports_kwh = 0
    annual_grid_exports_kwh = 0

    for s in seasons:
        for t in range(5):
            # Calculate energy for a single representative day
            daily_gen = generation[s][t] * optimal_C_PV * degradation_factor
            daily_load = demand[s][t]
            net_energy = daily_gen - daily_load
            
            # Since BESS is 0, surplus is exported, deficit is imported
            if net_energy > 0:
                annual_grid_exports_kwh += days[s] * net_energy
            else:
                annual_grid_imports_kwh += days[s] * abs(net_energy)
            
            annual_generation_kwh += days[s] * daily_gen

    # Calculate costs and savings for the year
    cost_of_imports = annual_grid_imports_kwh * grid_buy_price
    revenue_from_exports = annual_grid_exports_kwh * grid_sell_price
    net_bill_with_pv = cost_of_imports - revenue_from_exports
    annual_savings = baseline_annual_bill - net_bill_with_pv
    
    results.append({
        "Year": year,
        "Annual PV Generation (kWh)": round(annual_generation_kwh, 2),
        "Net Bill with PV (₹)": round(net_bill_with_pv, 2),
        "Annual Savings (₹)": round(annual_savings, 2)
    })

# --- Display Results in a Table ---
df_results = pd.DataFrame(results)
print("Projected Annual Electricity Bills and Savings")
print("---------------------------------------------")
print(f"Baseline Annual Bill (Grid-Only): ₹{baseline_annual_bill:,.2f}")
print(df_results.to_string(index=False))
```

```python
import pulp

# --- 1. Define Problem Data and Parameters ---
# These values are taken from the assignment document.
seasons = ["Monsoon", "Dry"]
days = {"Monsoon": 122, "Dry": 243}
time_blocks = range(5)
demand = {"Monsoon": [0.8, 1.3, 1.1, 2.6, 0.5], "Dry": [0.8, 1.6, 1.3, 2.4, 0.5]}
generation = {"Monsoon": [0.2, 1.5, 1.0, 0.0, 0.0], "Dry": [0.7, 4.0, 3.0, 0.0, 0.0]}

# Financial and technical parameters
discount_rate = 0.08
pv_capex = 65000
bess_capex = 28000
pv_area_per_kwp = 100
grid_buy_price = 9
grid_sell_price = 5
bess_efficiency = 0.90
initial_budget = 300000
max_rooftop_area = 500

# NPV factors
pvaf_25_8 = (1 - (1 + discount_rate)**-25) / discount_rate
pvf_10_8 = (1 + discount_rate)**-10

# New Subsidy Parameters from the problem description
subsidy_per_kw = 15000
subsidy_cap_kw = 4.0

# --- 2. Create the MILP Model ---
model = pulp.LpProblem("Home_Energy_System_with_Subsidy", pulp.LpMinimize)

# --- 3. Define Decision Variables ---
# Sizing variables
C_BESS = pulp.LpVariable("C_BESS", lowBound=0, cat='Continuous')

# Variables to handle the capped subsidy for the PV system
C_PV_sub = pulp.LpVariable("C_PV_sub", lowBound=0, cat='Continuous') # Subsidized portion
C_PV_nosub = pulp.LpVariable("C_PV_nosub", lowBound=0, cat='Continuous') # Non-subsidized portion
b = pulp.LpVariable("b", cat='Binary') # Binary switch

# Derived variable for total PV capacity
C_PV = C_PV_sub + C_PV_nosub

# Operational variables (unchanged)
keys = [(s, t) for s in seasons for t in time_blocks]
E_grid = pulp.LpVariable.dicts("E_grid", keys, lowBound=0)
E_export = pulp.LpVariable.dicts("E_export", keys, lowBound=0)
E_charge = pulp.LpVariable.dicts("E_charge", keys, lowBound=0)
E_discharge = pulp.LpVariable.dicts("E_discharge", keys, lowBound=0)
SOC = pulp.LpVariable.dicts("SOC", keys, lowBound=0)

# --- 4. Define the Updated Objective Function ---
# Pre-subsidy investment (used for O&M calculation)
pre_subsidy_investment = pv_capex * C_PV + bess_capex * C_BESS
# Net investment (with subsidy applied)
net_initial_investment = (pv_capex * C_PV) - (subsidy_per_kw * C_PV_sub) + (bess_capex * C_BESS)

# NPV components
bess_replacement_cost = bess_capex * C_BESS * pvf_10_8
npv_om_cost = 0.01 * pre_subsidy_investment * pvaf_25_8
annual_net_grid_cost = pulp.lpSum(days[s] * (grid_buy_price * E_grid[s, t] - grid_sell_price * E_export[s, t]) for s, t in keys)
npv_net_grid_cost = annual_net_grid_cost * pvaf_25_8

model += (net_initial_investment + bess_replacement_cost + npv_om_cost + npv_net_grid_cost), "Total_Lifecycle_Cost"

# --- 5. Define the Updated Constraints ---
model += net_initial_investment <= initial_budget, "Budget_Constraint"
model += pv_area_per_kwp * C_PV <= max_rooftop_area, "Rooftop_Area_Constraint"

# Subsidy cap logic constraints
big_M = 100 # A sufficiently large number
model += C_PV_sub <= subsidy_cap_kw, "Subsidy_Cap_Upper_Limit"
model += C_PV_sub >= subsidy_cap_kw * b, "Force_Full_Subsidy_If_Over"
model += C_PV_nosub <= big_M * b, "Activate_NonSubsidized_Portion"

# Energy balance and battery operation constraints (use total C_PV)
for s in seasons:
    for t in time_blocks:
        model += (generation[s][t] * C_PV + E_grid[s, t] + E_discharge[s, t]
                  == demand[s][t] + E_charge[s, t] + E_export[s, t]), f"Energy_Balance_{s}_{t}"
        t_prev = (t - 1 + 5) % 5
        model += (SOC[s, t] == SOC[s, t_prev] + bess_efficiency * E_charge[s, t]
                  - E_discharge[s, t]), f"SOC_Update_{s}_{t}"
        model += SOC[s, t] <= C_BESS, f"SOC_Capacity_{s}_{t}"
        model += E_discharge[s, t] <= SOC[s, t_prev], f"Discharge_Limit_{s}_{t}"
    model += (pulp.lpSum(bess_efficiency * E_charge[s, t] - E_discharge[s, t] for t in time_blocks)
              == 0), f"Daily_Cycle_{s}"

# --- 6. Solve the Problem and Print Results ---
model.solve(pulp.PULP_CBC_CMD(msg=0)) # Solve silently

# --- 7. Display the New Optimal Configuration ---
if pulp.LpStatus[model.status] == 'Optimal':
    print("Optimal Solution with Subsidy Found ✅")
    print("------------------------------------------")
    print(f"Size of the solar PV system: {C_PV.value():.2f} kWp")
    print(f"Size of the BESS: {C_BESS.varValue:.2f} kWh")
    print("------------------------------------------")
    print(f"(Breakdown: {C_PV_sub.varValue:.2f} kWp subsidized, {C_PV_nosub.varValue:.2f} kWp non-subsidized)")
else:
    print(f"Solver status: {pulp.LpStatus[model.status]}")
```

```python
import numpy_financial as npf

# --- NPV Calculation Breakdown in Python ---

# Inputs
pv_size = 5.0      # kWp
bess_size = 1.25   # kWh
pv_capex = 65000   # Rs/kWp
bess_capex = 28000   # Rs/kWh
subsidy_cap = 4.0    # kWp eligible
subsidy_rate = 15000 # Rs/kWp
discount_rate = 0.08
lifetime = 25
bess_efficiency = 0.90
grid_buy_price = 9
grid_sell_price = 5

# --- 1. Net Initial Investment (Year 0 Cost) ---
gross_pv_cost = pv_size * pv_capex
gross_bess_cost = bess_size * bess_capex
total_gross = gross_pv_cost + gross_bess_cost

subsidy = min(pv_size, subsidy_cap) * subsidy_rate
net_investment = total_gross - subsidy

# --- 2. NPV of BESS Replacement (Year 10 -> start of year 11) ---
replacement_year = 10
replacement_cost = gross_bess_cost
discount_factor_10 = (1 / (1 + discount_rate) ** replacement_year)
npv_replacement = replacement_cost * discount_factor_10

# --- 3. NPV of O&M Costs ---
annual_om = 0.01 * total_gross
# Present value annuity factor for 25 years at 8%
pvaf_25 = (1 - (1 + discount_rate) ** -lifetime) / discount_rate
npv_om = annual_om * pvaf_25

# --- 4. NPV of Net Grid Revenue (given result from simulation/code) ---
npv_grid_revenue = 479807  # Rs, positive revenue

# --- 5. Final NPV Calculation ---
npv_total = net_investment + npv_replacement + npv_om - npv_grid_revenue

# --- 6. Payback Period Calculation ---
# Energy profile data needed for payback calculation
seasons = ["Monsoon", "Dry"]
days = {"Monsoon": 122, "Dry": 243}
demand = {"Monsoon": [0.8, 1.3, 1.1, 2.6, 0.5], "Dry": [0.8, 1.6, 1.3, 2.4, 0.5]}
generation = {"Monsoon": [0.2, 1.5, 1.0, 0.0, 0.0], "Dry": [0.7, 4.0, 3.0, 0.0, 0.0]}

# a) Calculate Baseline Annual Bill (Grid-Only)
annual_demand_kwh = sum(days[s] * sum(demand[s]) for s in seasons)
baseline_annual_bill = annual_demand_kwh * grid_buy_price

# b) Calculate Net Bill with PV & BESS (Year 1)
annual_net_grid_revenue_yr1 = 0
for s in seasons:
    for d in range(days[s]):
        soc = 0 # Assume battery starts empty each day
        for t in range(5):
            solar_gen = generation[s][t] * pv_size # Year 1, no degradation
            load = demand[s][t]
            
            # 1. Discharge battery to meet load if possible
            discharge_amount = min(load, soc * bess_efficiency)
            load -= discharge_amount
            soc -= discharge_amount / bess_efficiency
            
            net_energy = solar_gen - load
            
            if net_energy > 0: # Surplus
                # 2. Charge battery with surplus
                charge_amount = min(net_energy, (bess_size - soc))
                soc += charge_amount
                # 3. Export the rest
                export = net_energy - charge_amount
                annual_net_grid_revenue_yr1 += export * grid_sell_price
            else: # Deficit
                # 4. Import from grid if needed
                annual_net_grid_revenue_yr1 += net_energy * grid_buy_price

net_bill_with_pv_yr1 = -annual_net_grid_revenue_yr1

# c) Calculate First-Year Net Savings
gross_savings_yr1 = baseline_annual_bill - net_bill_with_pv_yr1
net_savings_yr1 = gross_savings_yr1 - annual_om

# d) Calculate Simple Payback Period
payback_period = net_investment / net_savings_yr1 if net_savings_yr1 > 0 else float('inf')

# --- Output ---
print("--- NPV Calculation Breakdown ---")
print(f"Net Initial Investment: ₹{net_investment:,.2f}")
print(f"NPV of BESS Replacement: ₹{npv_replacement:,.2f}")
print(f"NPV of O&M: ₹{npv_om:,.2f}")
print(f"NPV of Net Grid Revenue: ₹{npv_grid_revenue:,.2f}")
print(f"Final NPV (25 years): ₹{npv_total:,.2f}")
print("\n--- Simple Payback Period Calculation ---")
print(f"Baseline Annual Bill (Grid-Only): ₹{baseline_annual_bill:,.2f}")
print(f"Net Bill with System (Year 1): ₹{net_bill_with_pv_yr1:,.2f}")
print(f"First-Year Gross Savings: ₹{gross_savings_yr1:,.2f}")
print(f"Annual O&M Cost: ₹{annual_om:,.2f}")
print(f"First-Year Net Savings: ₹{net_savings_yr1:,.2f}")
print(f"Simple Payback Period: {payback_period:.2f} years")
```
